```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("scrape_data.csv")
df.head(5)
```

```python
df.info()
```

```python
df.drop(["radio"], axis=1, inplace=True)
df.columns
```

```python
df['price'] = df['price'].str.replace(r'€', '')
df['price'] = df['price'].str.replace(r',', '')
df['price'] = df['price'].str.replace(r'.', '')
df['price'] = df['price'].str.replace(r' ', '')
df['price'][0]
```

```python
kw_power = []
for power in df["power"]:
    if isinstance(power, str):
        p, _, _, _= power.split(" ")
        kw_power.append(p)
    else:
        kw_power.append(power)

df["power"] = kw_power
```

```python
df['mileage'] = df['mileage'].str.replace(r'km', '')
df['mileage'] = df['mileage'].str.replace(r',', '')
df['mileage'] = df['mileage'].str.replace(r' ', '')
df['mileage'][0]
```

```python
fr_year = []
for year in df["first registration"]:
    if isinstance(year, str):
        _, p = year.split("/")
        fr_year.append(p)
    else:
        fr_year.append(year)
df["first registration"] = fr_year
```

```python
df['equipment'] = df['equipment'].str.replace('\n', '')
df['equipment'] = df['equipment'].str.split(',')
df['equipment']
```

```python
df.head()
```

```python
df.to_csv('clean_data.csv')
```
